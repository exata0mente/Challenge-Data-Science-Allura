#### Junção das bases

Após dividir e consquistar, vou mesclar as bases em um único dataframe.

Nesta mesclagem irei utilizar a função ```reduce()``` do pacote ```functools``` pois ela é uma alternativa para mesclar múltiplos dataframe, já que a função ```merge()``` do `pandas` aceita apenas dois df por chamada.

Usarei essa função ao inves da ```concat()``` do pacote ```pandas``` justamente pelo fato de assegurar que os dataframes serão mesclados por uma variável comum.

Para entender a função ```reduce()``` usei estas referências ([1](https://stackoverflow.com/questions/44327999/python-pandas-merge-multiple-dataframes), [2](https://pt.stackoverflow.com/questions/291837/pra-que-serve-a-fun%C3%A7%C3%A3o-reduce-em-python) em pt-br) 

````python
from functools import reduce


df_mesclar = [json_bruto[['customerID', 'Churn']], dados_customer, dados_phone, dados_internet, dados_account]

dados = reduce(lambda df_aux, df_lista: pd.merge(df_aux, 
                                                 df_lista, 
                                                 left_index=True, 
                                                 right_index=True, 
                                                 how='inner'),
               df_mesclar)
dados.head()
````

````python
dados.columns
````

### Tradução das colunas

````python
nome_colunas = {
    'customerID': 'id_cliente',
    'Churn': 'fl_inativo',
    'gender': 'sexo',
    'SeniorCitizen': 'fl_maior_65_anos',
    'Partner': 'fl_tem_parceiro',
    'Dependents': 'fl_dependentes',
    'tenure': 'meses_contrato',
    'PhoneService': 'fl_servico_telefonico',
    'MultipleLines': 'fl_multiplas_linhas',
    'InternetService': 'tipo_servico_internet',
    'OnlineSecurity': 'fl_servico_seguranca_online',
    'OnlineBackup': 'fl_servico_backup_online', 
    'DeviceProtection': 'fl_servico_protecao_dispositivo',
    'TechSupport': 'fl_servico_suporte_tech',
    'StreamingTV': 'fl_servico_tv_cabo',
    'StreamingMovies': 'fl_servico_stream',
    'Contract': 'tipo_contrato',
    'PaperlessBilling': 'fl_sem_papel',
    'PaymentMethod': 'tipo_pagamento',
    'Charges.Monthly': 'valor_servico_mensal' ,
    'Charges.Total': 'valor_servico_total'
}
````

````python
dados.columns.shape[0] == len(nome_colunas)
````

### Tradução de conteúdo

````python
traducao_flag = {'Yes' : 'Sim', 'No': 'Não', 0 : 'Não', 1: 'Sim'}
colunas_flag = dados.filter(like='fl_').columns
colunas_flag
````

````python
for coluna in colunas_flag:
    dados[coluna].replace(traducao_flag, inplace=True)
````

````python
colunas_traducao = {'Male': 'Masculino', 'Female': 'Feminino'}
dados.sexo.replace(colunas_traducao, inplace=True)
````
